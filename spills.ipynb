```python
# Date Range
df['Spill Date'] = pd.to_datetime(df['Spill Date'])
df['Year'] = df['Spill Date'].dt.year
years = df['Year'].value_counts()
print(df['Year'].dtypes)
print(years.sort_index(ascending=True))
```

```python
# Group by the 'county' column and sum the 'spill_value' for each county
#county_spill_sums = df.groupby('Material Name')['Quantity'].sum()

# Display the result
#print(county_spill_sums)
```

```python
# Creating a Recovered Amount Column
df['Unrecovered Amount'] = df['Quantity'] - df['Recovered']

#Function for Getting Unrecovered Amount
def unrecovered(row):
     amt = row['Quantity'] - row['Recovered']
     if amt < 0:
        amt = 0
     return amt

df['Unrecovered Amount'] = df.apply(unrecovered, axis=1)
address_df = df.copy()
```

```python
df = df.dropna(subset=['Program Facility Name'])
con_ed_df = df[df['Program Facility Name'].str.contains('CON ED')]
con_ed_df = con_ed_df[con_ed_df['Unrecovered Amount'] > 0]
pd.set_option('display.max_rows', None)
con_ed_df.sort_values(by=['Year'], ascending=False)
```

```python
# SELECTING AND FILTERING
condition1 = df['Material Family'] == 'Hazardous Material' 
condition2 = df['Material Family'] == 'Petroleum'
condition3 = df['Year'] >= 2020

hazMat_county_sums = df[condition1 & condition3].groupby('County')['Unrecovered Amount'].sum()
print(hazMat_county_sums.sort_values(ascending=False))
```

```python
#UNIQUE VALUE COUNTS
condition1 = df['Year'] >= 2020
source = df[condition1]
source['Source'].value_counts().sort_values(ascending=False)
```

```python
cond = df['Year'] >= 2020
contr_factor = df[cond]
contr_factor['Contributing Factor'].value_counts().sort_values(ascending=False)
```

```python
#DOUBLE GROUPBY
gb_cols = ['Source', 'Contributing Factor']
df.groupby(gb_cols)['Unrecovered Amount'].sum()
```

```python
address_df['Unrecovered Amount'] = df.apply(unrecovered, axis=1)
address_df = address_df[address_df['Year'] >= 2024]
address_df = address_df[address_df['Material Family'].isin(['Petroleum', 'Hazardous Material'])]

address_df.drop(columns=['Spill Number', 'ZIP Code', 'SWIS Code', 'DEC Region', 'Received Date', 'Close Date'], inplace=True)
address_df.sort_values(by=['Unrecovered Amount'], ascending=False).head(10)
```
