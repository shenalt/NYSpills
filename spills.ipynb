```python
df_24 = df_24.dropna(subset=['Program Facility Name'])
print('DF 24')
print(df_24.isnull().sum())
print('-------------------------')
df_25 = df_25.dropna(subset=['Program Facility Name'])
print('DF 25')
print(df_25.isnull().sum())
```

```python
# Date Range
df_24['Spill Date'] = pd.to_datetime(df_24['Spill Date'])
df_24['Year'] = df_24['Spill Date'].dt.year
years = df_24['Year'].value_counts()
print('DF 24')
print(df_24['Year'].dtypes)
print(years.sort_index(ascending=True))

print('DF 25')
df_25['Spill Date'] = pd.to_datetime(df_25['Spill Date'])
df_25['Year'] = df_25['Spill Date'].dt.year
years = df_25['Year'].value_counts()
print(df_25['Year'].dtypes)
print(years.sort_index(ascending=True))
```

```python
# Creating a Recovered Amount Column
df_24['Unrecovered Amount'] = df_24['Quantity'] - df_24['Recovered']
df_25['Unrecovered Amount'] = df_25['Quantity'] - df_25['Recovered']

#Function for Getting Unrecovered Amount
def unrecovered(row):
     amt = row['Quantity'] - row['Recovered']
     if amt < 0:
        amt = 0
     return amt

df_24['Unrecovered Amount'] = df_24.apply(unrecovered, axis=1)
df_25['Unrecovered Amount'] = df_25.apply(unrecovered, axis=1)
df_25
```

```python
#con_ed_df = df[df['Program Facility Name'].str.contains('CON ED')]
#con_ed_df = con_ed_df[con_ed_df['Unrecovered Amount'] > 0]
#pd.set_option('display.max_rows', None)
#con_ed_df.sort_values(by=['Year'], ascending=False)

merged_df = pd.merge(df_25, df_24, on='Spill Number', how='left')
merged_df = merged_df.drop(columns=['Unrecovered Amount_y', 'Year_y', 'Recovered_y', 'Units_y', 'Quantity_y', 'Material Family_y', 'Material Name_y', 'Close Date_y', 'Source_y', 'Waterbody_y',
                                    'Contributing Factor_y', 'Received Date_y', 'Spill Date_y', 'DEC Region_y', 'SWIS Code_y', 'ZIP Code_y', 'County_y', 'Locality_y', 'Street 2_y', 'Street 1_y',])
merged_df.columns
```

```python
con_ed_df = merged_df[(merged_df['Program Facility Name_x'].str.contains('CON ED')) | (merged_df['Program Facility Name_y'].str.contains('CON ED'))]
con_ed_df.sort_values(by=['Year_x'], ascending=False).head(40)
```

```python
# SELECTING AND FILTERING
condition1 = con_ed_df['Unrecovered Amount_x'] > 99 
condition2 = con_ed_df['Year_x'] >= 2000

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

con_ed_df = con_ed_df[condition1 & condition2]
con_ed_df.sort_values(by=['Year_x'], ascending=False)
```
